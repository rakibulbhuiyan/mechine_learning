```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('car driving risk analysis.csv',sep=',')
dataset=pd.read_csv('data.csv',sep=',')
```

```python
dataset
```

```python
dataset.columns
```

```python
x = dataset.iloc[:, :-1].copy()

# Display the new DataFrame
x
```

```python
y=dataset.drop(dataset.columns[:-1], axis=1)

# Display the new DataFrame
y
```

```python
#taking care of missing Data

# missing_values=dataset.isnull().sum()
mean_age = dataset['Age'].mean()
mean_salary = dataset['Salary'].mean()
dataset['Age'].fillna(mean_age, inplace=True)
dataset['Salary'].fillna(mean_salary, inplace=True)
```

```python
dataset
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.NaN, strategy='mean')
imputer.fit(x.iloc[:, 1:3])
x.iloc[:, 1:3] = imputer.transform(x.iloc[:, 1:3])
x
```

```python
imputer = SimpleImputer(missing_values=np.NaN, strategy='mean')
x_values = x.values  # Convert DataFrame to numpy array
imputer.fit(x_values[:, 1:3])
x_values[:, 1:3] = imputer.transform(x_values[:, 1:3])
x = pd.DataFrame(x_values, columns=x.columns)  # Convert back to DataFrame
x
```

```python
dataset
```

```python
# Using get_dummies on the 'Country' column
dummies = pd.get_dummies(dataset['Country'])

# Concatenating the dummy variables with the original DataFrame
dataset = pd.concat([dataset, dummies], axis='columns')
```

```python
dataset
```

```python
final=dataset.drop(['Country'],axis='columns')
```

```python
final
```

```python
final['France'] = final['France'].replace({True: 1, False: 0})
final['Germany'] = final['Germany'].replace({True: 1, False: 0})
final['Spain'] = final['Spain'].replace({True: 1, False: 0})
final['Purchased'] = final['Purchased'].replace({'Yes': 1, 'No': 0})
final
```

```python
y=final[['Purchased']]
```

```python
y
```

```python
x=final.drop(['Purchased'],axis=1)
```
